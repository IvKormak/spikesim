```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
dom = np.arange(-1, 1, 1/1000)
tanh = np.tanh(dom*np.pi)
step = (dom>0.5).astype(np.int32)+0.0070
sigma = 1/(1+np.exp(-dom*5))
lin = dom
relu = dom*(dom>0)+0.003
lrelu = 0.9*dom*(dom>0)+0.1*dom
fig, ax = plt.subplots(1,2, figsize=(10,4))
ax[1].set_yticks((0, 1))
ax[0].set_yticks((-1, 0, 1), (-1, 0, 1))

for _ax in ax:
    _ax.spines['left'].set_position(('data', 0))
    _ax.spines['bottom'].set_position('zero')
    _ax.spines['right'].set_visible(False)
    _ax.spines['top'].set_visible(False)
    
    _ax.set_xticks(())
    
    _ax.spines.bottom.set_bounds(-1, 1)
    
    
    _ax.set_xlabel("z", loc="right")
ax[0].set_xticks((0.5,), ("$z_{th}$",))
ax[0].plot(
    dom, tanh,
    dom, step,
    dom, lin,
)
ax[1].plot(
    dom, lrelu,
    dom, relu,
    dom, sigma,
)
plt.savefig("графики.png")
```
